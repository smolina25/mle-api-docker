# JSON

**JSON**, or **JavaScript Object Notation**, is a **lightweight data interchange format** that has become the **standard for data exchange between applications**. It was originally developed for use with **JavaScript**, but it's now widely used across different **programming languages and platforms**.

**JSON data** is represented in a **text format** and consists of **key-value pairs** or **arrays of data**, similar to **Python dictionaries**. These key-value pairs are enclosed in **curly braces**, and each pair is separated by a **comma**. The **key** is always a **string**, and the **value** can be a **string, number, boolean, null, array, or another JSON object**.

One of the **main advantages of JSON** is that it's **easy to read and write** for both humans and machines. It's also a **platform-independent format**, which means it can be used across different programming languages and platforms without any issues.

**JSON** is widely used in **web development**, especially for **exchanging data between a web server and a client**. It's also used in **mobile app development**, **IoT devices**, and many other applications where **data needs to be exchanged between different systems**.  

**JSON is a popular format for data exchange in APIs** due to its **simplicity, flexibility, and ease of use**. Most modern APIs use JSON as their **primary data exchange format**.  

When a **client** sends a **request** to an **API endpoint**, it typically sends **JSON data** as part of the **request body**, which the API can then **parse and use to process the request**. Similarly, when the **API returns a response** to the client, it usually does so in **JSON format**, which the client can **easily parse and use to display the data**.  

**APIs that use JSON** as their **primary data exchange format** are **easy to integrate with other systems and platforms**, making them a **popular choice for building modern applications**.  
<br>

Let's look at an **example** of a JSON object:

```json
{
    "name": "John Doe",
    "age": 30,
    "email": "john.doe@example.com",
    "address": {
        "street": "123 Main St",
        "city": "Anytown",
        "state": "CA",
        "zip": "12345"
    },
    "phoneNumbers": [
        {
            "type": "home",
            "number": "555-555-1234"
        },
        {
            "type": "work",
            "number": "555-555-5678"
        }
    ],
    "isEmployee": true,
    "salary": null
}
```

In this example, we see how a **JSON object** can contain **nested objects** and **arrays**.  

The **outermost object** contains a **list of key-value pairs**, where the **key** is a **string** and the **value** can be a **string, number, boolean, null, array, or another JSON object**.  

The **`address`** key contains another **JSON object**, which in turn contains a **list of key-value pairs**.  

Similarly, the **`phoneNumbers`** key contains an **array of JSON objects**, each of which contains a **list of key-value pairs**.  

We can also **import this JSON file into a Python dictionary** using the **`json` module**:  

````python
import json
````

### Serializing JSON

What happens after a computer processes lots of information? It needs to take a **data dump**.  

Accordingly, the **`json` library** exposes the **`dump()`** method for **writing data to files**.  

There is also a **`dumps()`** method (pronounced as *“dump-s”*) for **writing to a Python string**.  

**Simple Python objects** are **translated to JSON** according to a fairly **intuitive conversion**.  

Python | JSON
---|---
dict | object
list, tuple | array
str | string
int, float, int- & float-derived Enums | number
True | true
False | false
None | null

An example of the **`dump()`** method:

````python
example_dict = {
    "name": "John Doe",
    "age": 30,
    "email": "john.doe@example.com",
    "address": {
        "street": "123 Main St",
        "city": "Anytown",
        "state": "CA",
        "zip": "12345"
    },
    "phoneNumbers": [
        {
            "type": "home",
            "number": "555-555-1234"
        },
        {
            "type": "work",
            "number": "555-555-5678"
        }
    ],
    "isEmployee": True,
    "salary": None
}
````

Using **Python’s context manager**, you can **create a file** called **`example.json`** and **open it in write mode** (JSON files conveniently end in a **`.json`** extension).  

````python
with open('data/example.json', 'w') as f:
    json.dump(example_dict, f)
````

Note that **`dump()`** takes **two positional arguments**:  

1. The **data object** to be **serialized**  
2. The **file-like object** to which the **bytes will be written**  

Or, if you were so inclined to continue using this **serialized JSON data** in your program, you could write it to a native Python **`str` object**.  

````python
json_string = json.dumps(example_dict)
````

````python
print(json_string)
````

Notice that the **file-like object is absent** since you aren’t actually writing to disk. Other than that, **`dumps()`** is just like **`dump()`**.

### Deserializing JSON

The **`json` library** also exposes the **`load()`** method for **reading data from files**.  

There is also a **`loads()`** method for **reading from a Python string**.  

JSON | Python
---|---
object | dict
array | list
string | str
number (int) | int
number (real) | float
true | True
false | False
null | None

Technically, this **conversion** isn’t a perfect **inverse** to the **serialization table**. That basically means that if you **encode an object now** and then **decode it again later**, you may not get exactly the **same object** back. It’s probably like getting **one friend to translate something into Japanese** and **another friend to translate it back into English**.  

Regardless, the simplest example would be **encoding a tuple** and getting back a **list after decoding**, like so:  

````python
blackjack_hand = (8, "Q")
encoded_hand = json.dumps(blackjack_hand)
decoded_hand = json.loads(encoded_hand)

print(blackjack_hand)
print(decoded_hand)
````

````python
blackjack_hand == decoded_hand
````

````python
type(blackjack_hand)
````

````python
type(decoded_hand)
````

````python
blackjack_hand == tuple(decoded_hand)
````

But let's look at the **`load()`** method, we will load the data that we dump earlier.

````python
# import the data from the JSON file which is in the directory data/example.json
with open('data/example.json') as f:
    persons = json.load(f) 
    
# print the data
print(persons)
````

If you want to **extract a specific value** from the JSON object, you can use the **```get()```** method:


````python
print(persons.get('name'))
````

If you want to **change the value of a specific key**, you can use the **```update()```** method:

````python
persons.update({"name": "Jane Doe"})
print(persons.get("name"))
````

If you want to **add a new person** to the JSON object, you can use the **```append()```** method.

We have the **original JSON string**:

````python
json_string = '''
[
    {
        "name": "John Doe",
        "age": 30,
        "email": "john.doe@example.com",
        "address": {
            "street": "123 Main St",
            "city": "Anytown",
            "state": "CA",
            "zip": "12345"
        },
        "phoneNumbers": [
            {
                "type": "home",
                "number": "555-555-1234"
            },
            {
                "type": "work",
                "number": "555-555-5678"
            }
        ],
        "isEmployee": true,
        "salary": null
    }
]
'''
````

And the **person we want to add as dictionary**:

````python
new_person = {
    "name": "Jane Doe",
    "age": 25,
    "email": "jane.doe@example.com",
    "address": {
        "street": "456 Main St",
        "city": "Anytown",
        "state": "CA",
        "zip": "12345"
    },
    "phoneNumbers": [
        {
            "type": "home",
            "number": "555-555-4321"
        }
    ],
    "isEmployee": False,
    "salary": None
}
````

Then we can **load the original JSON string with ```json.loads()```** and **add the person** to the JSON string:

````python
persons = json.loads(json_string)
persons.append(new_person)
print(persons)
````

Now that we have two people in this JSON file we can use **slicing** to get the first person:

````python
persons[0]
````

or the second person:

````python
persons[1].get('name')
````

We used **pydantic** earlier this week to validate a **pandas dataframe**, but **JSON** is where **pydantic really shines**.  

When working with **JSON data**, pydantic's **data models** can be used to **define the structure** of the JSON object. Pydantic's **parsing** and **validation features** ensure that the JSON data is **well-formed** and matches the **expected schema**.  

This makes it easy to handle JSON data in Python, as pydantic can automatically **serialize** and **deserialize** the JSON data to and from **Python objects** that match the defined **data model**.  

Let's look at an example. If we have a JSON like this:

```json
{
    "name": "John Doe",
    "age": 30,
    "email": "john.doe@example.com",
    "address": {
        "street": "123 Main St",
        "city": "Anytown",
        "state": "CA",
        "zip": "12345"
    },
    "phone": "555-555-1234"
}
```

We can write a **pydantic model** to **validate the JSON**, but because we have a **nested JSON `address`**, we need to create a **pydantic model** for the nested JSON as well and then we use it as a **type hint** for the **`address` key** in the **main model**.  

> You can find a nice way to create those pydantic models [**here**](https://jsontopydantic.com/).

````python
from pydantic import BaseModel, ValidationError

class Address(BaseModel):
    street: str
    city: str
    state: str
    zip: str

class User(BaseModel):
    name: str
    age: int
    email: str
    address: Address
    phone: str
````

Now we can **use the pydantic model to validate the JSON**:

````python
response_string= '''{
    "name": "John Doe",
    "age": 30,
    "email": "john.doe@example.com",
    "address": {
        "street": "123 Main St",
        "city": "Anytown",
        "state": "CA",
        "zip": "12345"
    },
    "phone": "555-555-1234"
}'''
````

You can use the **```model_validate_json```** method to **parse the JSON** into a pydantic model (~~**```parse_raw```**~~ method is now deprecated) or you can use the **```model_validate```** method to **parse a python dictionary** into a pydantic model (~~**```parse_obj```**~~ is also deprecated):

````python
try:
    # Parse the JSON string into a Person object
    person = User.model_validate_json(response_string)
    print(person)
except ValidationError as e:
    # Handle validation errors
    print(e)
````

````python
try:
    # Parse the JSON string into a Person object
    response_json = json.loads(response_string)
    person = User.model_validate(response_json)
    print(person)
except ValidationError as e:
    # Handle validation errors
    print(e)
````
